```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\T L S\\Desktop\\SoftSinc\\Week 2\\IRIS.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='species', order=df['species'].value_counts().index)
plt.title('Category-wise Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df.reset_index(),  
    x='index',                    
    y='sepal_length',
    hue='species',                  
    marker='o'
)

plt.title('Sepal Length Trend Over Rows by Species')
plt.xlabel('Row Index')
plt.ylabel('Sepal Length')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='sepal_length', bins=20, kde=True)
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
corr = df[numeric_cols].corr()


sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x='species', y=col)
    plt.title(f'Boxplot of {col} by Species')
    plt.xlabel('Species')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```
